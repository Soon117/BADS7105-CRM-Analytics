```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import matplotlib.pyplot as plt
import networkx as nx
df = pd.read_csv('Customer Preference Survey (Responses).csv')
```

```python
df
```

```python
#Drop first row
df = df.drop(0)
df
```

```python
#drop duplicated PS5 column
df_clean = df.drop(columns=['PS5', 'Timestamp'])
df_clean
```

```python
#fillna
df_clean = df_clean.fillna(0)
df_clean
```

```python
#แปลงเคยซื้อ ไม่เคยซื้อเป็น 0,1
df_clean=df_clean.replace('ไม่เคยซื้อ',0)
df_clean=df_clean.replace('ไม่เคย',0)
df_clean=df_clean.replace('ไม่',0)
df_clean=df_clean.replace('ไม่เคยซือ',0)
df_clean=df_clean.replace('เคยซื้อ',1)
df_clean=df_clean.replace('เคย',1)
#df_clean.iloc[1].unique()
df_clean
```

```python
basket_encode_final_df = df_clean[(df_clean>0).sum(axis=1) >= 2]
basket_encode_final_df
```

```python
freq_items_set = apriori(basket_encode_final_df,
                        min_support=0.1,
                        use_colnames=True).sort_values('support', ascending=False).reset_index(drop=True)

freq_items_set
```

```python
rules = association_rules(freq_items_set, metric="lift", min_threshold=1).sort_values('lift', ascending=False).reset_index(drop=True)

rules['consequents_lenght'] = rules['consequents'].apply(lambda x : len(x))
rules['antecedents_lenght'] = rules['antecedents'].apply(lambda x : len(x))

rules
```

```python
rules[(rules['antecedents_lenght'] <= 1) &
      (rules['consequents_lenght'] <= 1) &
      (rules['lift'] >= 2) &
      (rules['support'] >= 0.1) &
      (rules['confidence'] >= 0.5) ]
```

```python
item_item_matrix = pd.DataFrame(index=basket_encode_final_df.columns,columns=basket_encode_final_df.columns)
```

```python
from scipy.spatial.distance import cosine

for i in range(0,len(item_item_matrix.columns)) :
    # Loop through the columns for each column
    for j in range(0,len(item_item_matrix.columns)) :
      # Fill in placeholder with cosine similarities
      item_item_matrix.iloc[i,j] = 1 -cosine(basket_encode_final_df.iloc[:,i],basket_encode_final_df.iloc[:,j])
```

```python
links = item_item_matrix.rename_axis('related item', 
                                     axis='columns').stack().reset_index()
links.columns = ['item', 'related item','value']
```

```python
links_filtered=links.loc[ (links['value'] > 0.1) & 
                         (links['item'] != links['related item']) ]
```

```python
import matplotlib.pyplot as plt
import networkx as nx

fig, ax=plt.subplots(figsize=(10,4))
GA=nx.from_pandas_edgelist(links_filtered,source='item',target='related item')
nx.draw(GA,with_labels=True)
plt.show()
```
